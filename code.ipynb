```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
df = pd.read_csv('/Users/damodargupta/Desktop/spam classification using naive bayes/spambase/spambase.data' )
df
```

```python
df.isnull().sum()
```

```python
# independent and dependent variables
X = df.iloc[: , :-1]
y = df.iloc[: , -1]
```

```python
X
```

```python
y
```

```python
y.value_counts()
```

# spam (1) or not (0)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Bernoulli classifier

```python
classifier = BernoulliNB()
classifier.fit(X_train , y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(classification_report(y_pred , y_test))
print(confusion_matrix(y_pred ,y_test))
print(accuracy_score(y_pred,y_test))
```

# Multinomial classifier

```python
classifier = MultinomialNB()
classifier.fit(X_train , y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(classification_report(y_pred , y_test))
print(confusion_matrix(y_pred ,y_test))
print(accuracy_score(y_pred,y_test))
```

# Gaussian Classifier

```python
classifier = GaussianNB()
classifier.fit(X_train , y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(classification_report(y_pred , y_test))
print(confusion_matrix(y_pred ,y_test))
print(accuracy_score(y_pred,y_test))
```

### in case of Bernoulli naive bayes classifier , we obtained an accuracy of 88.21 %

### in case of Multinomial naive bayes classifier , we obtained an accuracy of 78.53 %

### in case of Gaussian naive bayes classifier , we obtained an accuracy of 82.29 %

Thus, we get the best results in bernoulli classifier and we know that it works best in case of binary classification problems such as this problem where the two classes are the mail is a spam and it is not a spam
